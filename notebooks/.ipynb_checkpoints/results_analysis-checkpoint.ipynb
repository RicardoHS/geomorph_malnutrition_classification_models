```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols_to_show = ['Accuracy', 'Specificity', 'Precision', 'Recall', 'F1']
cols_to_show_extra = cols_to_show + ['total_SAM', 'correct_SAM', ' total_MAM', 'correct_MAM', ' total_RIS', 'correct_RIS', ' total_NOR', 'correct_NOR']
```

```python
results_path = !ls ../results/results*
```

```python
results_path
```

```python
df_all = pd.DataFrame()

for result_file in results_path:
    df = pd.read_csv(result_file)
    df = df.rename(columns={'Unnamed: 0':'name'})
    df['model'] = df['name'].str.split('->').apply(lambda x: x[-1]).astype('category')
    df['method'] = df['name'].str.split('->').apply(lambda x: '->'.join(x[:-1])).astype('category')
    df['experiment'] = result_file.replace('../results/results_','').split('.')[-2]
    df['experiment'] = df['experiment'].str.split('_').apply(lambda x: '_'.join(x[-2:]))
    
    extra_cols_indicator = result_file.replace('../results/results_','').split('-')
    if len(extra_cols_indicator) > 1:
        df['extra_cols'] = 'sex+age_group'
    elif extra_cols_indicator[0][:3] == 'age':
        df['extra_cols'] = 'age_group'
    elif extra_cols_indicator[0][:3] == 'sex':
        df['extra_cols'] = 'sex'
    else:
        df['extra_cols'] = ''
        
    df_all = pd.concat([df_all, df])
```

# Independent results

```python
for result_file in results_path:
    df = pd.read_csv(result_file)
    df = df.rename(columns={'Unnamed: 0':'name'})
    df['model'] = df['name'].str.split('->').apply(lambda x: x[-1]).astype('category')
    df['method'] = df['name'].str.split('->').apply(lambda x: '->'.join(x[:-1])).astype('category')

    df.sort_values('Accuracy', ascending=False).set_index('name')['Accuracy'].plot.bar(ylim=[0,1], figsize=(17,7), title='Accuracy by method + model: ' +result_file.split('/')[-1])
    plt.show()
```

```python
for result_file in results_path:
    df = pd.read_csv(result_file)
    df = df.rename(columns={'Unnamed: 0':'name'})
    df['model'] = df['name'].str.split('->').apply(lambda x: x[-1]).astype('category')
    df['method'] = df['name'].str.split('->').apply(lambda x: '->'.join(x[:-1])).astype('category')

    df.groupby('method').mean().sort_values('Accuracy', ascending=False)['Accuracy'].plot.bar(ylim=[0,1], figsize=(17,7), title='Mean Accuracy by method: ' + result_file.split('/')[-1])
    plt.show()
```
